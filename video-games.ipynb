```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
pd.options.display.max_columns = 50
df = pd.read_csv('./sales.csv')
df = df.drop(axis=1, labels='index')
df['Platform'] = df['Platform'].astype('category')
```

## Find the best selling genre per each region (all time)

```python
genres_df = df.groupby('Genre')[['Europe', 'Japan', 'Rest of World', 'North America', 'Global']]
sales_sum_per_genre = (genres_df
.sum()
)
best_genre = sales_sum_per_genre.idxmax()
best_genre
```

## Genre sales per region (all time)

```python
fig, axes = plt.subplots(nrows=5, ncols=1)
for idx, column in enumerate(sales_sum_per_genre.columns):

  sales_sum_per_genre[column].sort_values().plot(kind='barh', figsize=(15,30), ax=axes[idx])
  axes[idx].set_xlabel('Sales (Millions of units)')
  axes[idx].set_ylabel('Genre')
  axes[idx].set_title(('Game sales in ' + column + ' by genre'))
  axes[idx].grid(True, axis='x')
  gridIntervalLength = math.ceil(sales_sum_per_genre[column].max()) /10
  xticks = np.arange(0, gridIntervalLength * 11, gridIntervalLength, dtype=int)
  axes[idx].set_xticks(xticks)
```

## Game releases in years

```python
df['Year'] = pd.to_numeric(df['Year'], downcast='integer')
releasesPerYear = df['Year'].value_counts().sort_index()
plot = (releasesPerYear
    .plot(kind='line',
      figsize=(15,5)))
plot.set_xlabel('Year')
plot.set_ylabel('Game releases')
plot.set_title('Number of game releases per year.')
```

### Most successful platform of all time (by average rank)

```python
platformGroups = df.groupby('Platform')[['Rank']].mean().sort_values(by='Rank')
platformGroups
```

```python
#remove all games from platforms with few games
#create a dataframe where indexes are platforms and columns the avg rank through years
#display a chart
yearRanges = [
    (1983, 1988),
    (1988, 1993),
    (1993, 1998),
    (1998, 2003),
    (2003, 2008),
    (2008, 2013)
]
rankInYears = pd.DataFrame()

# rankInYears.head()
popularPlatformsList = df.groupby('Platform')[['Rank']].mean().sort_values(by='Rank').iloc[0:10].index
popularPlatforms = df[df['Platform'].isin(popularPlatformsList)].reset_index()[['Platform', 'Rank', 'Year']].sort_values(by='Rank')
for range in yearRanges:
    columnName = 'Avg_Rank_' + str(range[0]) + '-' + str(range[1])
    rankInYears[columnName] = popularPlatforms.query("@range[0] <= Year < @range[1]").mean()
rankInYears.head()
```
